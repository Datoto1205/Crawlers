# Crawler of Top 10 headlines of Wall Street Journal with Chinese Version 

```python
from urllib import request
from bs4 import BeautifulSoup
import requests

numberOfNews = 1

# Check the status of the target website
html = request.urlopen('https://wallstreetcn.com/news/global?uid=109176')
print(html)
#In python3, urllib2 module was seperated into several different sub-modules, so we need to import it with more complicated way.

resp = requests.get('https://wallstreetcn.com/news/global?uid=109176') # Return an object with request.Response type.
print("The status of the server of the target website is: " + str(resp.status_code))
# The status code of an object means whether we have obtained response from the server, 200 means yes, but 404 means error.
```

```python
# Examine the whole HTML code of target website
soup = BeautifulSoup(resp.text, 'html.parser')
HTMLCodeOfWebsite = soup.prettify()
print(HTMLCodeOfWebsite)

# Fetch the data
newsTitle = soup.find_all('a', class_ = 'title')
print(newsTitle)
```

```python
# Sort the data
for tag in newsTitle:
    if numberOfNews < 11:
        print("The " + str(numberOfNews) + " news: " + tag.text)
        print("The link is: https://wallstreetcn.com/" + str(tag.get('href')) + "\n")
        numberOfNews += 1
```
